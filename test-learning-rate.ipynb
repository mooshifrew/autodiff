```python
%load_ext autoreload 
%autoreload 2
from matplotlib import pyplot as plt
import pickle as pkl
import numpy as np
from autodiff.activation import ReLU, Linear
from autodiff.network import Network, NetworkParams

file_path = 'test-parameters.pkl'
plt.style.use("tableau-colorblind10")

with open(file_path, 'rb') as file:
    data = pkl.load(file)
```

```python
def get_loss(y, y_hat): 
    return 0.5*((y_hat - y)**2)
```

```python
network_def: NetworkParams = {
    "input_shape": 2,
    "output_shape": 1,
    "layers": [
        {
            "input_shape": 2,
            "n_neurons": len(data['w1']),
            "weight_init": data['w1'],
            "bias_init": data['b1'] ,
            "activation": ReLU(),
        },
        {
            "input_shape": len(data['w1']),
            "n_neurons": len(data['w2']),
            "weight_init": data['w2'],
            "bias_init": data['b2'] ,
            "activation": ReLU(),
        },
        {
            "input_shape": len(data['w2']),
            "n_neurons": len(data['w3']),
            "weight_init": data['w3'],
            "bias_init": data['b3'] ,
            "activation": Linear(), 
        }
    ]
} 
```

```python
network = Network(network_def)
epochs = 100 # so that the loss after the final update is calculated
loss_vals1 = []
learning_rate = 1
batch_size = len(data['inputs'])
for iteration in range(epochs): 
    print(f'EPOCH: {iteration}')
    current_loss = 0
    network.zero_grad()
    for index, input in enumerate(data['inputs']):
        y_pred = network.forward(input)
        y = data['targets'][index] 
        loss = get_loss(y, y_pred)
        current_loss += loss
        network.backward(y_pred - y)


    print(f"    Average loss: {current_loss/batch_size}")
    network.update_params(learning_rate=learning_rate, batch_size=batch_size)
    loss_vals1.append(current_loss/batch_size)
```
